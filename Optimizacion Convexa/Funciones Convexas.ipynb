# **Funcion Convexa** 

Una funcion $f: \mathbb{R}^n \rightarrow \mathbb{R}$ es **convexa** si el **dominio** de la funcion $f$ es un **conjunto convexo** y si para cualquier dos puntos $x, y$ pertenecientes al dominio se cumple que: 

$$f(\theta x + (1 - \theta) y) \leq \theta f(x) + (1 - \theta) f(y) \hspace{0.8cm} 0 \leq \theta \leq 1$$

Una funcion $f: \mathbb{R}^n \rightarrow \mathbb{R}$ es **estrictamente convexa** si el **dominio** de la funcion $f$ es un **conjunto convexo** y si para cualquier dos puntos distintos $x, y$ pertenecientes al dominio se cumple que: 

$$f(\theta x + (1 - \theta) y) < \theta f(x) + (1 - \theta) f(y) \hspace{0.8cm} 0 < \theta < 1 \hspace{0.8cm} x \neq y$$

La **desigualdad** de convexidad significa que, el **segmento de linea** que une los dos puntos $x, y$ en el **dominio** de $f$ siempre esta por **encima o toca** el grafico de $f$

**Nota:** Si la funcion $f$ es **convexa** o **estrictamente convexa**, la funcion $-f$ es **concava** o **estrictamente concava**, Las **funciones afines y lineales** son **Convexas y Concavas** al mismo tiempo.

```python
# Definimos los Valores de la Constante que Multiplica al Exponente

a_values = [-7, 1, 3]

# Definimos el Dominio de la Funcion

x = np.linspace(-2, 2, 400)

# Definimos los Puntos donde se definira el Segmento de Linea 

x1, x2 = -0.7, 1.5

# Visualizamos 

fig, axes = plt.subplots(1, 3, figsize = (15, 4))

for i, a  in enumerate(a_values):
    
    f = lambda x: np.exp(a * x)
    y = f(x)
    
    y1, y2 = f(x1), f(x2)
    t = np.linspace(0, 1, 100)
    x_line = t * x1 + (1 - t) * x2
    y_line = t * y1 + (1 - t) * y2

    ax = axes[i]
    ax.plot(x, y, color = 'blue')
    ax.plot(x_line, y_line, 'r--', label = 'Segmento de Linea')
    ax.scatter([x1, x2], [f(x1), f(x2)], color = 'black')

    ax.set_title(rf'$f(x) = e^{{{a}x}}$')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.axhline(0, color = 'black', linewidth = 0.5)
    ax.axvline(0, color = 'black', linewidth = 0.5)
    ax.grid(True)
    ax.legend()

plt.tight_layout()
plt.show()
```

# **Restriccion de una Funcion Convexa a una Linea** 

Una funcion $f: \mathbb{R}^n \rightarrow \mathbb{R}$ es **convexa** si y solo si la funcion $g: \mathbb{R} \rightarrow \mathbb{R}$ es **convexa en t** para todo $x$ perteneciente al **dominio de $f$**

$$g(t) = f(x + tv) \hspace{0.8cm} \text{dom} \hspace{0.05cm} g = \{ t \mid x + tv \in \text{dom} \hspace{0.05cm} f \} \hspace{0.8cm} v \in \mathbb{R}^n$$

**Nota:** Esto nos indica que en todas las **posibles direcciones** la funcion $g(t)$ tiene que ser **convexa** para que la funcion $f$ sea convexa

### **Convexa**

```python
# Definimos el Dominio de la Funcion

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x, y)

# Definimos la Funcion

Z = X**2 + Y**2

# Visualizamos

fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(111, projection = '3d')

ax.plot_surface(X, Y, Z, cmap = 'BrBG', alpha = 0.9)

ax.set_title(r'$f(x, y) = x^2 + y^2$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.tight_layout()
plt.show()
```

```python
# Definimos un Punto de Estudio dentro del Dominio de la Funcion Convexa

x0 = np.array([1.0, -1.0])

# Definimos Multiples Direcciones

v_values = [np.array([2.0, 1.0]), np.array([-7.0, 5.0]), np.array([3.5, -4.2]), np.array([8.1, 0.3])]

# Definimos el Dominio de T (Tiene que estar restringido t | x0 + vt en dom f) (En este caso no esta restringido)

t = np.linspace(-2, 2, 200)

# Definimos la Funcion

f = lambda x,y: x**2 + y**2

# Visualizamos 

fig, axes = plt.subplots(1, 4, figsize = (15, 4))

for i, v  in enumerate(v_values):
    
    x_t = x0[0] + t * v[0]
    y_t = x0[1] + t * v[1]
    
    g_t = f(x_t, y_t)
    
    ax = axes[i]
    ax.plot(t, g_t, color = 'blue')

    ax.set_title(rf'$g(t) = f(({x0[0]}, {x0[1]}) + t({v[0]}, {v[1]}))$')
    ax.set_xlabel('t')
    ax.set_ylabel('g(t)')
    ax.axhline(0, color = 'black', linewidth = 0.5)
    ax.axvline(0, color = 'black', linewidth = 0.5)
    ax.grid(True)

plt.tight_layout()
plt.show()
```

# **Extension de Valor** 

Es conveniente ampliar el dominio de una funcion **convexa** para todos los valores de $\mathbb{R}^n$, aadiendo $\infty$ para los valores fuera del **dominio** de $f$.

$$\tilde{f}(\theta x + (1 - \theta) y) \leq \theta \tilde{f}(x) + (1 - \theta) \tilde{f}(y) \hspace{0.8cm} 0 \leq \theta \leq 1$$

Convertimos el dominio de $f: \mathbb{R}^n \rightarrow \mathbb{R}$, a $\tilde{f}: \mathbb{R}^n \rightarrow \mathbb{R} \cup \infty$.

$$
\tilde{f}(x) = 
\begin{cases}
f(x) & \text{si } x \in \operatorname{dom} f \\
\infty & \text{si } x \notin \operatorname{dom} f
\end{cases}
$$

**Nota:** De manera similar extendemos las funciones **concavas** aadiendo $-\infty$ para los valores fuera del **dominio** de $f$.

### **Convexa**

# **Condicion de Primer Orden** 

Sea $f$ una funcion **diferenciable** una vez, $f$ es **convexa** si solo si, el **dominio** de la funcion $f$ es **convexo** y si se cumple lo siguiente para todos los $x, y$ pertenecientes al **dominio** de $f$: 

$$f(y) \geq f(x) + \nabla f(x)^T (y - x)$$

La **aproximacion de primer orden** en el cualquier punto de estudio $y$ que pertenezca al **dominio** de $f$, **No sobrepasa la grafica de la funcion en ningun punto**

**Nota:** Si el $\nabla f(x) = 0$ esta cota indica que $f(y) \geq f(x)$ para todo $y$, lo que nos dice que $x$ es un **minimo global**

### **Convexa**

```python
# Definimos la Funcion

f = lambda x: x**2

# Definimos el Gradiente de la Funcion 

df = lambda x: 2*x

# Definimos el Dominio y la Imagen de la Funcion

x = np.linspace(-3, 3, 500)
y = f(x)

# Puntos de Estudios

x0 = 1.5 
y0 = 1.9

# Calculamos la Condicion de Primer Oden con la Aproximacion de Taylor (Recta Tangente)

x_tangent = np.linspace(x0 - 0.75, x0 + 0.75, 100)

f_approx = lambda x: f(x0) + df(x0)*(x - x0)

y_tangent = f_approx(x_tangent)

# Visualizamos 

plt.plot(x, y, color = 'blue')
plt.plot(x_tangent, y_tangent, 'r--', label = 'Recta Tangente en $x_0$')
plt.scatter([x0], [f(x0)], color = 'black', label = f'$x_0$')
plt.scatter([y0], [f(y0)], color = 'green', label = f'$f(y_0) = {f(y0):0.6f}$')
plt.scatter([y0], [f_approx(y0)], color = 'red', label = f'Aprox. lineal en $y_0$ = {f_approx(y0):0.6f}')

plt.title(rf'$f(x) = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color = 'black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# **Condicion de Segundo Orden** 

Sea $f$ una funcion **diferenciable** dos veces, $f$ es **convexa** si solo si, el **dominio** de la funcion $f$ es **convexo** y si se cumple lo siguiente para todos los $x$ pertenecientes al **dominio** de $f$: 

$$\nabla^2 f(x) \succeq 0 \hspace{0.8cm} \text{Hessiana semidefinida positiva}$$

La **aproximacion de segundo orden** en el cualquier punto de estudio $y$ que pertenezca al **dominio** de $f$, **Su curvatura local siempre es positiva o nula**

### **Convexa**

```python
# Definimos la Funcion

f = lambda x: np.exp(3 * x)

# Definimos el Gradiente de la Funcion

df = lambda x: 3 * np.exp(3 * x)

# Definimos la Hessiana de la Funcion

df2 = lambda x: 9 * np.exp(3 * x)

# Definimos el Dominio y la Imagen de la Funcion

x = np.linspace(-3, 3, 500)
y = f(x)

# Puntos de Estudios

x0 = 1.7
y0 = 2.2

# Calculamos la Condicion de Primer Orden con la Aproximacion de Taylor (Recta Tangente)

x_tangent = np.linspace(x0 - 0.75, x0 + 0.75, 100)

f_approx = lambda x: f(x0) + df(x0)*(x - x0)

y_tangent = f_approx(x_tangent)

# Calculamos la Condicion de Segundo Orden con la Aproximacion de Taylor (Curvatura)

x_curvature = np.linspace(x0 - 0.75, x0 + 0.75, 100)

f2_approx = lambda x: f(x0) + df(x0)*(x - x0) + 0.5 * df2(x0) * (x - x0)**2

y_curvature = f2_approx(x_curvature)

# Visualizamos 

plt.plot(x, y, color = 'blue')
plt.plot(x_curvature, y_curvature, 'r--', label = 'Curvatura en $x_0$')
plt.scatter([x0], [f(x0)], color = 'black', label = f'$x_0$')
plt.scatter([y0], [f(y0)], color = 'green', label = f'$f(y_0) = {f(y0):0.6f}$')
plt.scatter([y0], [f2_approx(y0)], color = 'red', label = f'Curvatura en $y_0$ = {f2_approx(y0):0.6f}')

plt.title(r'$f(x) = e^{3x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color = 'black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# **Conjuntos de Subniveles**

El **conjunto de subniveles $\alpha$** de una funcion **convexa** $f: \mathbb{R}^n \rightarrow \mathbb{R}$, son tambien **convexos** para cualquier valor $\alpha$

$$C_{\alpha} = \{x \in \operatorname{dom} f \mid f(x) \leq \alpha\}$$

El **conjunto de superniveles $\alpha$** de una funcion **concava** $f: \mathbb{R}^n \rightarrow \mathbb{R}$, son **convexos** para cualquier valor $\alpha$

$$C_{\alpha} = \{x \in \operatorname{dom} f \mid f(x) \geq \alpha\}$$

**Nota:** Una funcion puede tener todos los **subniveles convexos** pero no necesariamente la funcion tiene que ser **convexa**

### **Convexa**

```python
# Definimos los Subniveles

alphas = [0.8, 1.5, 2.7, 3.5]

# Definimos el Dominio de la Funcion

x = np.linspace(-4.5, 4.5, 400)

# Visualizamos 

fig, axes = plt.subplots(1, 4, figsize = (15, 4))

for i, a in enumerate(alphas):
    
    f = lambda x: np.abs(x)**1
    y = f(x)
    
    ax = axes[i]
    
    x_sub = x[y <= a]
    y_sub = y[y <= a]
    
    ax.plot(x, y, color = 'blue')
    ax.axhline(a, color = 'red', linestyle = '--', label = fr'$\alpha = {a}$')
    ax.fill_between(x_sub, y_sub, a, where = (y_sub <= a), color = 'blue', alpha = 0.3)

    ax.set_title(r'$f(x) = |x|^{1}$')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.axhline(0, color = 'black', linewidth = 0.5)
    ax.axvline(0, color = 'black', linewidth = 0.5)
    ax.grid(True)
    ax.legend()

plt.tight_layout()
plt.show()
```

# **Epigraph**

El **grafico** de una funcion $f: \mathbb{R}^n \rightarrow \mathbb{R}$ se define como: 

$$\{(x, f(x)) \mid x \in \operatorname{dom} f\}$$

El **Epigraph** de una funcion $f: \mathbb{R}^n \rightarrow \mathbb{R}$ se define como: 

$$\operatorname{epi} f = \{(x, t) \mid x \in \operatorname{dom} f, \hspace{0.1cm} f(x) \leq t\}$$

El **Hypograph** de una funcion $f: \mathbb{R}^n \rightarrow \mathbb{R}$ se define como: 

$$\operatorname{hypo} f = \{(x, t) \mid x \in \operatorname{dom} f, \hspace{0.1cm} t \leq f(x)\}$$

**Nota:** La relacion entre **conjuntos convexos** y **funciones convexas** es mediante el **epigraph**. Una **funcion es convexa** si solo si, su **epigraph** es un **conjunto convexo**

### **Convexa**

```python
# Definimos la Funcion

f = lambda x: x**2

# Definimos el Gradiente de la Funcion 

df = lambda x: 2*x

# Definimos el Dominio y la Imagen de la Funcion

x = np.linspace(-4, 4, 500)
y = f(x)

# Puntos de Estudios

x0 = 2 
y0 = 2

# Calculamos la Condicion de Primer Oden con la Aproximacion de Taylor (Recta Tangente)

x_tangent = np.linspace(x0 - 2, x0 + 2, 100)

f_approx = lambda x: f(x0) + df(x0)*(x - x0)

y_tangent = f_approx(x_tangent)

# Vector Normal a la Recta Tangente (Hiperplano de Soporte en x0)

m = df(x0)

normal = np.array([m, -1])
normal = normal / np.linalg.norm(normal)

scale = 1.0 

# Visualizamos 

plt.plot(x, y, color = 'blue')
plt.plot(x_tangent, y_tangent, 'r--', label = 'Recta Tangente en $x_0$')
plt.text(-0.2, 6, r'$\operatorname{epi} f$', fontsize = 15, color = 'black')

for i in range(len(x)):
    plt.vlines(x[i], y[i], y[i] + 20, color = 'skyblue', alpha = 0.1)
    
plt.scatter([x0], [f(x0)], color = 'black', label = f'$x_0$')
plt.scatter([y0], [f(y0)], color = 'green')
plt.scatter([y0], [f_approx(y0)], color = 'red')
plt.quiver(*np.array([x0, f(x0)]), *normal*scale, angles = 'xy', scale_units = 'xy', scale = 1, color = 'purple',  label = 'Vector normal')

plt.title(rf'$f(x) = x^2$')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 16)
plt.axhline(0, color = 'black', linewidth = 0.5)
plt.axvline(0, color = 'black', linewidth = 0.5)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
